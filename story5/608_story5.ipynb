```python
import pandas as pd

url = "https://raw.githubusercontent.com/Naik-Khyati/608/main/story5/input/temp_hurricanes.csv"
data = pd.read_csv(url)
```

```python
data.info()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Extract data from the DataFrame
years = data['Year']
nbr_hur_adj = data['nbr_hur_adj']
temp_anomaly = data['temp_anamoly']
spdi = data['spdi']

# Define color-blind neutral colors
colors = ['#986eac', '#3a93c3', '#f6a582']

# Create line charts for each variable
plt.figure(figsize=(12, 6))

# Line chart for nbr_hur_adj with a label
plt.plot(years, nbr_hur_adj, label='Hurricanes Count', marker='o', linestyle='-', color=colors[0])

# Line chart for temp_anomaly
plt.plot(years, temp_anomaly, label='Temperature Anomaly', marker='o', linestyle='-', color=colors[1])

# Line chart for spdi
plt.plot(years, spdi, label='Power Dissipation Index', marker='o', linestyle='-', color=colors[2])

# Customize the plot
#plt.title('Climate and Hurricane Data Over the Years')
plt.xlabel('Year')
plt.ylabel('Values')
plt.grid(True)

# Highlight the periods with strong correlation
strong_correlation_periods = [(1980, 1984), (1988, 1992), (1996, 2000), (2004, 2008), (2012, 2016)]
for period in strong_correlation_periods:
    plt.axvspan(period[0], period[1], color='gray', alpha=0.2)

# Create legends for all three variables and "Periods of Strong Correlation"
legend_elements = [
    plt.Line2D([0], [0], color=colors[0], label='Hurricanes Count'),
    plt.Line2D([0], [0], color=colors[1], label='Temperature Anomaly'),
    plt.Line2D([0], [0], color=colors[2], label='Power Dissipation Index'),
    plt.Line2D([0], [0], color='gray', alpha=0.2, label='Strong Correlation Periods')
]

plt.legend(handles=legend_elements)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Extract data from the DataFrame
years = data['Year']
nbr_hur_adj = data['nbr_hur_adj']
temp_anomaly = data['temp_anamoly']
spdi = data['spdi']

# Standardize the variables
nbr_hur_adj_std = (nbr_hur_adj - nbr_hur_adj.mean()) / nbr_hur_adj.std()
temp_anomaly_std = (temp_anomaly - temp_anomaly.mean()) / temp_anomaly.std()
spdi_std = (spdi - spdi.mean()) / spdi.std()

# Define color-blind neutral colors
colors = ['#986eac', '#3a93c3', '#f6a582']

# Create line charts for each standardized variable
plt.figure(figsize=(12, 6))

# Line chart for standardized hurricanes count
plt.plot(years, nbr_hur_adj_std, label='Standardized Hurricanes Count', marker='o', linestyle='-', color=colors[0])

# Line chart for standardized temperature anomaly
plt.plot(years, temp_anomaly_std, label='Standardized Temperature Anomaly', marker='o', linestyle='-', color=colors[1])

# Line chart for standardized power dissipation index
plt.plot(years, spdi_std, label='Standardized Power Dissipation Index', marker='o', linestyle='-', color=colors[2])

# Customize the plot
#plt.title('Standardized Climate and Hurricane Data Over the Years')
plt.xlabel('Year')
plt.ylabel('Standardized Values')
plt.grid(True)

# Highlight the periods with strong correlation
strong_correlation_periods = [(1980, 1984), (1988, 1992), (1996, 2000), (2004, 2008), (2012, 2016)]
for period in strong_correlation_periods:
    plt.axvspan(period[0], period[1], color='gray', alpha=0.2)

# Create legends for all three standardized variables and "Periods of Strong Correlation"
legend_elements = [
    plt.Line2D([0], [0], color=colors[0], label='Standardized Hurricanes Count'),
    plt.Line2D([0], [0], color=colors[1], label='Standardized Temperature Anomaly'),
    plt.Line2D([0], [0], color=colors[2], label='Standardized Power Dissipation Index'),
    plt.Line2D([0], [0], color='gray', alpha=0.2, label='Strong Correlation Periods')
]

plt.legend(handles=legend_elements)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Extract data from the DataFrame
temp_anomaly = data['temp_anamoly']
nbr_hur_adj = data['nbr_hur_adj']

# Define color-blind neutral colors
colors = ['#986eac', '#3a93c3', '#f6a582']

# Create scatter plots for temp_anomaly vs. nbr_hur_adj
plt.figure(figsize=(12, 6))

# Scatter plot for temp_anomaly vs. nbr_hur_adj
plt.scatter(temp_anomaly, nbr_hur_adj, color=colors[0], label='Temperature vs. Hurricanes', marker='o')

# Calculate the correlation line for temp_anomaly vs. nbr_hur_adj using seaborn
sns.regplot(x=temp_anomaly, y=nbr_hur_adj, color='black', scatter=False, label='Correlation Line (Temp vs. Hurricanes)')

# Customize the plot
plt.title('Correlation Between Temperature Anomaly and Hurricanes')
plt.xlabel('Temperature Anomaly (°C)')
plt.ylabel('Number of Hurricanes (Adjusted)')
plt.grid(True)

# Create a custom legend
legend_elements = [plt.Line2D([0], [0], marker='o', color=colors[0], label='Temperature vs. Hurricanes', markersize=5),
                   plt.Line2D([0], [0], linestyle='-', color='black', label='Correlation Line (Temp vs. Hurricanes)')]

plt.legend(handles=legend_elements)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Extract data from the DataFrame
temp_anomaly = data['temp_anamoly']
nbr_hur_adj = data['nbr_hur_adj']
spdi = data['spdi']

# Define color-blind neutral colors
colors = ['#986eac', '#3a93c3', '#f6a582']

# Create scatter plots for temp_anomaly vs. spdi
plt.figure(figsize=(12, 6))

# Scatter plot for temp_anomaly vs. spdi
plt.scatter(temp_anomaly, spdi, color=colors[2], label='Temperature vs. SPDI', marker='o')

# Calculate the correlation line for temp_anomaly vs. spdi using seaborn
sns.regplot(x=temp_anomaly, y=spdi, color='black', scatter=False, label='Correlation Line (Temp vs. SPDI)')

# Customize the plot
plt.title('Correlation Between Temperature Anomaly and SPDI')
plt.xlabel('Temperature Anomaly (°C)')
plt.ylabel('SPDI (Power Dissipation Index)')
plt.grid(True)

# Create a custom legend
legend_elements = [plt.Line2D([0], [0], marker='o', color=colors[2], label='Temperature vs. SPDI', markersize=5),
                   plt.Line2D([0], [0], linestyle='-', color='black', label='Correlation Line (Temp vs. SPDI)')]

plt.legend(handles=legend_elements)

# Show the plot
plt.show()
```

```python
# Select the three variables of interest
selected_variables = data[['temp_anamoly', 'nbr_hur_adj', 'spdi']]

# Calculate the correlation matrix for the selected variables
correlation_matrix = selected_variables.corr()
```

```python
correlation_strength = correlation_matrix['temp_anamoly'].drop('temp_anamoly')

# Customize the x-axis labels
correlation_strength.index = ['hurricane occurrences', 'power dissipation index']

correlation_strength.plot(kind='bar', color='skyblue')
# plt.title('Correlation Strength with Temperature Anomaly')
plt.xlabel('Variable')
plt.ylabel('Correlation Strength')
plt.xticks(rotation=0)

# Add values of correlation index on the bars
for i, v in enumerate(correlation_strength):
    plt.text(i, v, f'{v:.2f}', ha='center', va='bottom')

plt.show()
```
